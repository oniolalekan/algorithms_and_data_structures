  # Question One: Check Duplicates in Tree

  Given the `root` of a binary tree, check whether it is contains a duplicate value. If a duplicate exists, return the duplicate value. If there are multiple duplicates, return the one with the closest distance to the root. If no duplicate exists, return -1.

  ## Examples

#### Starter Code for Question 1

```python
# Definition for a binary tree node.
from collections import defaultdict

class TreeNode(object):
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

def is_duplicate(root: TreeNode) -> int:
    def serialize(node, counter):
        if not node:
            return "#"
        left_serial = serialize(node.left, counter)
        right_serial = serialize(node.right, counter)
        serial = f"{node.val},{left_serial},{right_serial}"
        counter[serial] += 1
        return serial

    counter = defaultdict(int)
    serialize(root, counter)
    
    # Count duplicates (subtrees that appear more than once)
    return sum(1 for count in counter.values() if count > 1)
```

  # Question Two: Path to Leaves

  Given the `root` of a binary tree, return all root to leaf paths in any order.

  ## Examples

  ### Example 1

  ![](./images/q1_ex1.png)

  Input: `root = [1, 2, 2, 3, 5, 6, 7]` *What traversal method is this?*

  Output: [[1, 2, 3], [1, 2, 5], [1, 2, 6], [1, 2, 7]]

  ### Example 2

  ![](./images/q1_ex3.png)

  Input: `root = [10, 9, 7, 8]`

  Output: [[10, 7], [10, 9, 8]]

</details>

#### Starter Code for Question 2

```python
from typing import List

# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right

def bt_path(root: TreeNode) -> List[List[int]]:
    def dfs(node, path, result):
        if not node:
            return
        path.append(node.val)
        if not node.left and not node.right:
            result.append(list(path))  # Copy the path
        else:
            dfs(node.left, path, result)
            dfs(node.right, path, result)
        path.pop()  # Backtrack

    result = []
    dfs(root, [], result)
    return result
```

  # Question Three: Missing Number in Range
 
  You are given a list containing `n` integers in the range `[0, n]`. Return a list of numbers that are missing from the range `[0, n]` of the array. If there is no missing number, return -1. Note, all the integers in the list may not be unique.
  
  ## Examples

  ### Example 1

  Input: `lst = [0, 2]`

  Output: [1]

  ### Example 2

  Input: `lst = [5, 0, 1]`

  Output: [2, 3, 4]

  ### Example 3

  Input: `lst = [6, 8, 2, 3, 5, 7, 0, 1, 10]`

  Output: [4, 9]

</details>

#### Starter Code for Question 3


```python
from typing import List

def missing_num(nums: List[int]) -> int:
    n = len(nums)
    expected_sum = n * (n + 1) // 2
    actual_sum = sum(nums)
    return expected_sum - actual_sum
```

```python
# Your answer here
from typing import List

def move_zeros(nums: List[int]) -> List[int]:
    insert_pos = 0

    # Move non-zero elements to the front
    for num in nums:
        if num != 0:
            nums[insert_pos] = num
            insert_pos += 1

    # Fill the rest with zeros
    while insert_pos < len(nums):
        nums[insert_pos] = 0
        insert_pos += 1

    return nums
```


-   Explain why their solution works in your own words.


```python
# Your answer here

# Step 1: Shift Non-Zero Elements Forward
# It iterates through the list and copies all non-zero values to the front of the list, one by one, using an insert_pos pointer.
# This keeps their original relative order intact.

# Step 2: Fill the Remaining Spaces with Zeros
# Once all non-zero elements are placed at the beginning, the rest of the list (from insert_pos onward) is filled with zeros.


```
